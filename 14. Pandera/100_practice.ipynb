```python
def verify_dataframe(df_1: pandas.DataFrame, df_2: pandas.DataFrame) -> str:
    print("estas columnas no esta en df_2:")
    for column_1 in df_1.columns:
        if column_1 in df_2.columns:
            pass
        else:
            print(f"{column_1}", df_1[column_1].dtype)
    
    print("\nestas columnas no esta en df_1:")
    for column_2 in df_2.columns:
        if column_2 in df_1.columns:
            pass
        else:
            print(f"{column_2}", df_2[column_2].dtype)
```

```python
verify_dataframe(df_1=df_wired_coaxial, df_2=df_wired_coaxial_with_amplifier)
```

```python
schema = pandera.DataFrameSchema({
    'level_id': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'unit_id': pandera.Column(str, nullable=False, required=True),
    'distance_to_soti': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'distance_to_sots': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'distance_to_distribution_board': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'backbone_cable_id': pandera.Column(str, nullable=False, required=True),
    'lateral_net_cable_id': pandera.Column(str, nullable=False, required=True),
    'connector_id': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'derivator_id': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'backbone_index': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'derivator_index': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'splitter_attenuation': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'connector_attenuation': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'backbone_cable_attenuation_with_47mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'backbone_cable_attenuation_with_1350mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'lateral_net_cable_attenuation_with_47mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'lateral_net_cable_attenuation_with_1350mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'derivator_attenuation': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'derivator_step_attenuation': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'cumulative_level_derivator_step_attenuation': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'total_attenuation_with_47mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'total_attenuation_with_1350mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'amplifier_id': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'amplifier_index': pandera.Column(int, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'amplifier_power': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'total_attenuation_amplifiers_with_47mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True),
    'total_attenuation_amplifiers_with_1350mhz': pandera.Column(float, pandera.Check(lambda x: x >= 0), nullable=False, required=True)
    })
```
