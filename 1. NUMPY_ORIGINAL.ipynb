## conjuntos

```python
x = np.array([1,4,5,8,-9,-6,5,-3,9,5])
np.unique(x) #se evita elementos repetitivos
```

```python
x = np.array([1,4,5,8,-9,-6,5,-3,9,5])
y = np.array([1,2,3,4,5,6,-6,5,-3,9,5])

np.union1d(x,y) #imprime valores iguales
```

```python
np.intersect1d(x,y) #valores iguales
```

```python
np.setdiff1d(x,y) #elementos de x menos el de y
```

```python
np.setxor1d(x,y) #imprime los valores no iguales
```
