```python
import pandas
import matplotlib.patheffects as pa
import matplotlib.pyplot as plt
# from matplotlib.ticker import FuncFormatter
```

```python
finanzas = {
    'Fecha': ['2/1/2018', '3/1/2018', '4/1/2018', '5/1/2018', '6/1/2018'],
    'Elon Musk': [171.42, 164.67, 184.33, 176.85, 178.29],
    'Bill Gates': [176.85, 171.29, 184.33, 176.86, 167.60]
}
```

```python
excel = pandas.DataFrame.from_dict(finanzas)
excel
```

```python
excel.dtypes
```

```python
excel['Fecha'] = pandas.to_datetime(excel['Fecha'], format= '%Y-%m-%d')
```

```python
excel.set_index('Fecha',inplace=True)
excel
```

```python
excel.dtypes
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize = (15, 8))
ax.plot(excel['Elon Musk'], label='Elon Musk', lw=2, color = 'r',
        path_effects = [pa.SimpleLineShadow(), pa.Normal()])
ax.plot(excel['Bill Gates'], label='Bill Gates', lw=2, color = 'g', linestyle= '-.',
        path_effects = [pa.SimpleLineShadow(), pa.Normal()])
ax.legend(shadow=True, prop={'family':'Serif', 'size': 15})
ax.set_ylabel('USD', fontname='Serif', fontsize=20)
ax.set_xlabel('Time', fontname='Serif', fontsize=20)
ax.set_title('Elon Musk vs Bill Gates', fontname='Serif', fontsize=25)
```

```python
import pandas
import matplotlib.patheffects as pa
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
df = pandas.read_csv('./compra_curso.csv', sep=';')
df
```

```python
grafico = df[['Compras_del_curso', 'calificaciones']]
grafico
```

```python
grafico.describe()
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize = (15, 8))
ax.plot(grafico.Compras_del_curso, grafico.calificaciones, marker='o', 
        linestyle='-', markersize=9,lw=2, color = 'r', markeredgecolor = 'b',
        markeredgewidth= 1.5)
ax.set_ylabel('calificacione', fontname='fantasy', fontsize=20)
ax.set_xlabel('Compras_del_curso', fontname='fantasy', fontsize=20)
ax.set_title('ventas y calificaciones', fontname='fantasy', fontsize=35, va='bottom')

def formato(x, pos):
    return "%.0f en mil compras"%(x/1000)

ax.set_xticklabels(list(ax.get_xticks()), fontname='fantasy', fontsize=13)
ax.xaxis.set_major_formatter(FuncFormatter(formato))
ax.set_yticklabels(list(ax.get_xticks()), fontname='fantasy', fontsize=13)
ax.legend(shadow=True, prop={'family':'Serif', 'size': 15})
plt.show()
```
