```python
import pandas as pd
# Create a data frame
df = pd.DataFrame ({
        'Group':  np.array(["joel","carlos","juan","pedros","diego"]),
        'Value': np.array([12,14,11,16,14])
})

# Initialize a Figure and an Axes
fig, ax = plt.subplots()

# Fig size
fig.set_size_inches(6,5)

# Create horizontal bars
ax.bar(x=df.Group, height=df.Value, hatch = ".", edgecolor = "b", label = "edades");
ax.set_ylim(0,19)
for i, v in enumerate(df.Value):
    ax.text(i-0.1 , v + .4, str(v), color='black')

plt.legend(loc = "upper right")
# Add title
ax.set_title('A simple barplot');
```

# Pie Plot

```python
import numpy as np
import matplotlib.pyplot as plt

paises = np.array(["Per√∫", "Chile", "Ecuador", "Colombia", "Bolivia"])
poblacion = np.array([10000,12000,8000,15000,3000])
porcentaje = np.array([(i/poblacion.sum())*100 for i in poblacion])

fig, ax = plt.subplots();
fig.set_size_inches(8,8)
ax.pie(porcentaje, labels = paises, explode= [0.1,0,0,0,0], autopct= lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*poblacion.sum()), 
        startangle=50, colors = ["b","r","y","purple","g"], shadow= True)
plt.legend(loc="upper left")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Represent years in rows in the table
rows = ['2014', '2015', '2016', '2017', '2018','2019','2020']
# Represent battery rating in columns of the table
columns = ('vaso', 'laptop', 'TV', 'Cameras', 'microfonos','puertas')

# Number of units sold each year, each rating. e.g. 75 units of 7Ah batteris sold in 2011
data = [[75, 144, 114, 102, 108,15],
        [90, 126, 102,  84, 126,16],
        [96, 114,  75, 105, 135,200],
        [105, 90, 150,  90,  75,100],
        [90,  75, 135,  75,  90,300],
        [90,  75, 135,  75,  90,80],
        [90,  75, 135,  75,  90,80]]

# Define the range and scale for y-axis
values = np.arange(0, 600, 100)

# Specify the color map to be used
colors = plt.cm.OrRd(np.linspace(0, 0.5, len(rows)))
n_rows = len(data)
index = np.arange(len(columns)) + 0.3
bar_width = 0.5

# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns))
fig, ax = plt.subplots()
fig.set_size_inches(8,6)

# Plot bars and create text labels for the table
cell_text = []

for row in range(n_rows):
    plot = plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x) for x in y_offset])
    i=0
    for rect in plot:
        ax.text(rect.get_x() + rect.get_width()/2, y_offset[i],'%d' % int(y_offset[i]), 
                ha='center', va='bottom')
        i = i+1   

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

plt.ylabel("Unidades vendidas")
plt.title('numero vendidos por a√±o')

# No ticks on X-axis, as table below will cover the labels
plt.xticks([])

plt.show()
```

## Polar Plot

```python
dinero_planeado = [30, 25, 10, 20, 30]
dinero_gastado = [32, 23, 11, 15, 32]
areas = ["Planeamiento", "recursos humanos", "produccion", "secretaria","Operaci√≥n"]

plt.figure(figsize=(10,6))
plt.subplot(polar = True)

(lines,labels) = plt.thetagrids(range(0,360, int(360/len(areas))), (areas))

theta = np.linspace(0, 2 * np.pi, len(dinero_planeado))
plt.plot(theta, dinero_planeado)
plt.fill(theta, dinero_planeado, 'b', alpha=0.1)

plt.plot(theta, dinero_gastado)
#plt.fill(theta, dinero_gastado, 'orange', alpha=0.4)

plt.legend(labels=('Plan','Actual'),loc="lower right")
plt.title("Distribucion de gastos")
plt.show()
```
